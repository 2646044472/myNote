```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,random_state=233,stratify=y)
```

```python
from sklearn.model_selection import GridSearchCV    # CV就是交叉验证
from sklearn.neighbors import KNeighborsClassifier
```

```python
params = {
    'n_neighbors':[n for n in range(1,20)],
    'weights':['uniform','distance'],
    'p':[p for p in range(1,7)]
}
```

```python
grid = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=params,
    n_jobs=-1
)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

## 交叉验证

```python
from sklearn.model_selection import cross_val_score
```

```python
neigh = KNeighborsClassifier()
cv_scores = cross_val_score(neigh, x_train, y_train, cv = 5)
cv_scores
```

```python
best_score = -1
best_n = -1
best_weight = ''
best_p = -1

for n in range(1,20):
    for weight in ['uniform','distance']:
        for p in range(1,7):
            neigh = KNeighborsClassifier(
                n_neighbors = n,
                weights = weight, #'uniform'
                p = p
            )
            neigh.fit(x_train,y_train)

            score = neigh.score(x_test,y_test)
            if score > best_score:
                best_score = score
                best_n = n
                best_weight = weight
                best_p = p
                
best_score, best_n, best_weight, best_p
```

```python
best_score = -1
best_n = -1
best_weight = ''
best_p = -1
best_sv_scores = None

for n in range(1,20):
    for weight in ['uniform','distance']:
        for p in range(1,7):
            neigh = KNeighborsClassifier(
                n_neighbors = n,
                weights = weight, #'uniform'
                p = p
            )
            neigh.fit(x_train,y_train)
            cv_scores = cross_val_score(neigh, x_train, y_train, cv = 5)
            score = np.mean(cv_scores)
            if score > best_score:
                best_score = score
                best_n = n
                best_weight = weight
                best_p = p
                best_sv_scores = cv_scores
best_score, best_n, best_weight, best_p, best_sv_scores
```
