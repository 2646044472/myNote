```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w, b = 1.8, 2.5
np.random.seed(0)
x1 = np.random.rand(100) * 4
noise = np.random.randn(100)
x2 = w * x1 + b + noise
x = np.vstack([x1, x2]).T
x -= np.mean(x, axis=0)
x.shape
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
pca.fit(x)
```

```python
x_pca = pca.transform(x)
x_pca_inv = pca.inverse_transform(x_pca)

plt.scatter(x[:, 0], x[:, 1], s = 10)
plt.scatter(x_pca_inv[:, 0], x_pca_inv[:, 1], s = 10, c = 'r')
plt.show()
```

## 图像降噪

```python
from sklearn.datasets import load_digits
digits = load_digits()
x = digits.data
y = digits.target
```

```python
x.shape, y.shape
```

```python
def plot_top20_digits(x):
    for i in range(20):
        plt.subplot(4, 5, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(x[i].reshape(8, 8), cmap=plt.cm.gray_r, interpolation="nearest")
    plt.show()
```

```python
plot_top20_digits(x)
```

```python
np.random.seed(0)
x_noise = x + np.random.randn(x.shape[0], x.shape[1]) *3
```

```python
plot_top20_digits(x_noise)
```

```python
pca = PCA(0.6)
pca.fit(x_noise)
x_noise_pca = pca.transform(x_noise)
x_noise_inv = pca.inverse_transform(x_noise_pca)
```

```python
plot_top20_digits(x_noise_inv)
```
